Q2. Write a Python program to do arithmetical operations addition and division.?

```python
task = input("Enter 'add' for addition and 'division' for division.").lower()
a = int(input("Enter 1st number: "))
b = int(input("Enter 2nd number: "))
if task == 'add' :
    c = a + b 
    print(f"The sum is {c}.")
elif task == 'division' :
    c = round((a / b ), 2)
    print(f"The division is {c}")
    
```

Q3 Write a Python program to find the area of a triangle?

```python
base = int(input("Enter the base length in meters: "))
height = int(input("Enter the height in meters: "))
area_of_triangle = round(0.5*base*height)
print(f"The area of the triangle is {area_of_triangle} m^2.")
```

Q4. Write a Python program to swap two variables?

```python
a = input("Enter variable a's value: ")
b = input("Enter variable b's value: ")
c = a
a = b 
b = c
print(f"value of a is {a}")
print(f"value of a is {b}")
```
